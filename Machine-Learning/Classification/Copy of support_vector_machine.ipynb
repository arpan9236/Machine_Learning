# Support Vector Machine (SVM)

Our third classification ML model is **Support Vector Machine (SVM)** for classification or also known as **Support Vector Classification (SVC)**. We will work on the same dataset "Social_Network_Ads".

All the code cells are **exactly same** as it was in logistic regression and K-NN models **except** the code cell where we built and trained the model on the training set (actually the "class" for the SVC is different from that of logistic regression and K-NN models).

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```
print(X_train)
```

```
print(y_train)
```

```
print(X_test)
```

```
print(y_test)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```
print(X_train)
```

```
print(X_test)
```

## Training the SVM model on the Training set

```
from sklearn.svm import SVC   # Import the SVC class from sklearn.svm module. We could either take LinearSVC class (which directly build the linear SVM for classification model) or SVC class with a linear kernel. In the next section, we will have the topic kernel SVM where we will choose linear and different non-linear (e.g., RBF) kernels for SVM model.
classifier_SVC = SVC(kernel = 'linear', random_state = 0)   # Create an object named classifier_SVC of the SVC class. There are many parameters of the SVC class, most of which we will use default values. We have used linear kernel (in the next section, we will use non-linear kernels like RBF or Polynomial). In this line, SVM for classification model is built, but not trained yet on the dataset.
classifier_SVC.fit(X_train, y_train)   # In this line, we train the SVC model on the training set. From the object classifier_SVC we call the fit() method to train on the training set.
```

## Predicting a new result

```
print(classifier_SVC.predict(sc.transform([[30,87000]])))
```

## Predicting the Test set results

```
y_pred = classifier_SVC.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Making the Confusion Matrix

```
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

**So, the SVC model has accuracy 90% which is larger than taht of logistic regression (89%) but lesser than that of K-NN model (93%)**.

## Visualising the Training set results

```
from matplotlib.colors import ListedColormap

X_set, y_set = sc.inverse_transform(X_train), y_train

X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.5))

plt.contourf(X1, X2, classifier_SVC.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)

plt.title('SVM (Training set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

## Visualising the Test set results

```
from matplotlib.colors import ListedColormap

X_set, y_set = sc.inverse_transform(X_test), y_test

X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 0.25),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 0.5))

plt.contourf(X1, X2, classifier_SVC.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('salmon', 'dodgerblue')))

plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())

for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('salmon', 'dodgerblue'))(i), label = j)

plt.title('SVM (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```

**So, the SVC model has accuracy 90% which is larger than taht of logistic regression (89%) but lesser than that of K-NN model (93%)**.

Why did not the SVC beat the K-NN model?

This is because we have chosen the "linear" kernel which separates different categories/classes of the target variable $y$ by a decision boundary which is a straight line and thus it has constarint i.e., even if you rotate the strainght line by small amount you will always have some (more) incorrect predictions.

So, if the decision boundary is non-linear (which we will do later section by using non-linear kernel like RBF or polynomial etc.) and curved having lesser constraint, we can get better predictions and better accuracy. The K-NN classifier uses a non-linear decision boundary and hence gives better result.
